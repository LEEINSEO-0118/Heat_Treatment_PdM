# 데이터 품질지수

Gartner의 6가지 데이터 품질 지수
1. 완전성(Completeness) : 필수항목에 누락이 없어야 한다.  
2. 유일성(Uniqueness) : 데이터 항목이 중복되어서는 안된다.
3. 유효성(Validity) : 데이터 항목이 정해진 유효범위 및 도메인을 충족해야 한다.
4. 일관성(Consistency) : 데이터가 지켜야 할 구조, 값, 표현되는 형태가 일관되게 정의되고, 서로 일치해야 한다.
5. 정확성(Accuracy) : 실제 존재하는 객체의 표현 값이 정확하게 반영이 되어야 한다.
6. 무결성(Integrity) : 자료의 오류 없이 변화에 영향을 받지 않고 데이터의 유일성, 유효성, 일관성이 보호되어야 한다.

#### 데이터 라벨링
사분위수의 [+- 1.5 x IQR]을 기준으로 이상치를 정의  
각 행의 이상치 count에 threshold를 적용하여 최종 이상치 라벨링을 수행

-> 이상치 threshold를 정하는 것은 도메인 지식이 풍부한 전문가와 정하는 것이 더욱 효과적  
-> 프로젝트에서는 데이터 비율로 판단하여 설정  

```python
eda_data['outliers_cnt'] = 0

for col in eda_data.columns:
    quantile_1, quantile_3 =  eda_data[col].quantile(0.25), eda_data[col].quantile(0.75)
    iqr = quantile_3 - quantile_1
    threshold_1 = quantile_1 - (1.5 * iqr)
    threshold_2 = quantile_3 + (1.5 * iqr)
    outliers =  (eda_data[col] < threshold_1) | (eda_data[col] > threshold_2)
    eda_data.outliers_cnt += outliers.astype(int)
```

```python
outliers = eda_data.outliers_cnt.value_counts()
print(outliers)
```

```python
sns.lineplot(data=outliers)
```

```python
outliers = outliers.reset_index()
outliers['Cumulative'] = outliers.outliers_cnt.cumsum()
outliers['Cumulative_Percentage'] = outliers.Cumulative / len(eda_data)
pd.options.display.float_format = '{:.3f}'.format
outliers
```

```python
# 9개 이상의 이상치를 가지면 이상 데이터로 간주
over_8 = outliers[outliers['index'] > 8]

print(f'이상치 개수 : {over_8.outliers_cnt.sum()}\n이상치 비율 :{float(over_8.outliers_cnt.sum() / len(eda_data) * 100): .3f}')
```

데이터 당 이상치 변수가 8개 초과인 것을 이상 데이터로 분류했을 때, 그 수가 2717, 비율이 0.094%로 0.1% 보다 작은 수준

```python
anomaly = (eda_data.outliers_cnt > 8).astype(int)
eda_data['anomaly'] = anomaly
```

```python
print(f'전체 데이터 : {len(eda_data)}')
print(f'정상 데이터 : {len(eda_data[eda_data.anomaly == 0])}')
print(f'이상 데이터 : {eda_data.anomaly.sum()}')
```

# 데이터 분리 & 파일 생성

```python
eda_data
```

```python
from sklearn.model_selection import train_test_split

normal_data = eda_data[eda_data.anomaly == 0]
anomaly_data = eda_data[eda_data.anomaly == 1]

train_data, temp_data = train_test_split(normal_data, test_size=0.3, random_state=42, shuffle=False) # Train 70%
valid_data, temp_data = train_test_split(temp_data, test_size=2/3, random_state=42, shuffle=False) # Valid 10%
test_data, pipeline_data = train_test_split(temp_data, test_size=0.5, random_state=42, shuffle=False) # Test 10%, Pipeline 10%
```

```python
print(f'Total Normal Data : {len(normal_data)}')
print(f'- Train Data : {len(train_data)}({len(train_data)/len(normal_data)*100:.0f}%)')
print(f'- Valid Data : {len(valid_data)}({len(valid_data)/len(normal_data)*100:.0f}%)')
print(f'- Test Data : {len(test_data)}({len(test_data)/len(normal_data)*100:.0f}%)')
print(f'- Pipeline Data : {len(pipeline_data)}({len(pipeline_data)/len(normal_data)*100:.0f}%)')
print(f'--------------------------------')
print(f'Anomaly Data : {len(anomaly_data)}')
print(f'--------------------------------')
print(f'Total Data : {len(total_data)}')
```
